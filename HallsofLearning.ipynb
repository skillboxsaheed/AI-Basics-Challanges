```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

playlist = {'Song': ['Beanie','Karma','Red Button','PRIDE','Middle of the Ocean','Romantic Homicide','YKWIM?','Get You'],
            'Artist':['Chezile','Kali Uchis','Drake','Kendrick Lamar','Drake','d4vd','Yot Club','Daniel Caesar'],
            'Energy':[1,0.95,0.9,1,0.96,0.98,1,0.93],
            'Day_Played':['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday','Monday']}
df = pd.DataFrame(playlist)
print(df)

print("---------------------------------------------------------")
print(df.head())

print("---------------------------------------------------------")
top_artist = df['Artist'].value_counts().idxmax()
top_count = df['Artist'].value_counts().max()

print(f"Your top artist is {top_artist} with {top_count} listens.")

print("---------------------------------------------------------")
print("Summary statistics:\n", df.describe()) #Gives summary statistics
print("\n")

songs_per_day = df['Day_Played'].value_counts()

days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
songs_per_day = songs_per_day.reindex(days_order)


plt.figure(figsize=(6, 4))
songs_per_day.plot(kind='bar', color='skyblue', edgecolor='black')

plt.title('Number of Songs Played Per Day')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Songs')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()

print("---------------------------------------------------------")
print("I think I listen to music mostly on Monday because Monday is lock in day and I like to listen to music while I work")

print("---------------------------------------------------------")
plt.figure(figsize=(5, 5))  # Make the chart square

# Plot the pie chart
plt.pie(
    df['Energy'],                      # Energy values for slices
    labels=df['Song'],                 # Song names as labels
    autopct='%1.1f%%',                 # Show percentage of total energy
    startangle=140,                    # Rotate starting angle
    colors=plt.cm.viridis(np.linspace(0, 1, len(df)))  # Color variety
)

# Add title and format the chart
plt.title('Energy Distribution per Song')
plt.axis('equal')  # Keep the pie chart circular
plt.tight_layout()  # Prevent clipping
plt.show()
```
